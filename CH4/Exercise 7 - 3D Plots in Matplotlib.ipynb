# Exercise 7 - 3D Plots in Matplotlib

In this exercise we will demonstrate how to generate a scatter plot in three dimensions in Python.

Import pandas and matplotlib.  To enable 3D plotting will also need to import 
```from mpl_toolkits.mplot3d import Axes3D```

````python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

Read in the dataset and select the columns Sepal Length, Sepal Width and Petal Width.

````python
df = pd.read_csv('iris-data.csv')[['Sepal Length', 'Sepal Width', 'Petal Width']]
df.head()
````

Plot the data in three dimensions, use the argument projection='3d' to the add_subplot method to create the 3D plot.

````python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Sepal Length'], df['Sepal Width'], df['Petal Width']);
ax.set_xlabel('Sepal Length (mm)');
ax.set_ylabel('Sepal Width (mm)');
ax.set_zlabel('Petal Width (mm)');
ax.set_title('Expanded Iris Dataset');
````
